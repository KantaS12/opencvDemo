```python
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline
```

```python
def download_and_unzip(url, folder):
    print(f"Downloading and extracting assests...", end="")

    urlretrieve(url, folder)

    try:
        with ZipFile(folder) as z:
            z.extractall(os.path.split(folder)[0])

        print("done")

    except Exception as e:
        print("\nINvalid file.",e)

URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"
asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB3.zip")
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)
```

```python
img = cv2.imread("Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)
plt.imshow(img[:,:, ::-1])
plt.show()
```

Drawing a Line:

img = cv2.line(img, pt1, pt2, color[, thickness[,
lineType[, shift]]])

img:image where we draw the line
pt1:first (x,y) location of the line
pt2: second point
color: color of line which will be drawn

Optional:
thickness: default is 1
lineType: type of line. Default is 8 (8-connected line). Usually, cv2.LINE_AA is used for the lineType.

cv2.LINE_16 is not supported

```python
imageLine = img.copy()
cv2.line(imageLine, (200,100), (400,100), (0,255,255),thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imageLine[:,:, ::-1])
plt.show()
```

Drawing a Circle:

img = cv2.circle(img, center, radius, color[,
thickness[, lineType[, shift]]])

img: image on which we draw the line
center: center of the circle
radius: radius of the circle
color: color of the circle

optional:
thickness: thickeness of the circle (positive), else it's filled circle (negative)
lineType: Type of circle boundary. Exact same as lineType in cv2.line

```python
imageCircle = img.copy()
cv2.circle(imageCircle, (900,500), 100, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imageCircle[:, :, ::-1])
plt.show()
```

Drawing a Rectangle:

img = cv2.rectangle(img, pt1, pt2, color[,
thickness[, lineType[, shift]]])

img: image on which the rectangle is to be drawn
pt1: vertex of the rectangle (top-left vertex)
pt2: vetex of the rectangle opposite to pt1 (bottom-right vertex)

optional:
thickness: positive -> thickness, negative -> filled
lineType: type of circle boundary. Exact same as lineType in cv2.line

```python
imageRectangle = img.copy()
cv2.rectangle(imageRectangle, (500, 100), (700, 600), (255,0 , 255), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imageRectangle[:, :, ::-1])
plt.show()
```

Adding Text:

img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[,
bottomLeftOrigin]]])

img: image on which the text is written on
text: text string to be written
org: bottom-left corner of the text string of the image
fontFace: font type
fontScale: font scale factor that is multiplied by the font-specific base size
    if fontScale is negative, the text is upside down
color: Font color

Optional:
thickness: integer specifying line thickness for text, default is 1
lineType: default value is 8 which is 8-connected line. usually, cv2.LINE_AA is used for the lineType

```python
imageText = img.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0,255,0)
fontThickness = 2
cv2.putText(imageText, text, (200,700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)
plt.imshow(imageText[:, :, ::-1])
plt.show()
```
